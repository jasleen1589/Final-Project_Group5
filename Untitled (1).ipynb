```python
import pandas as pd

# Assuming your DataFrame is already loaded as df
# df = pd.read_csv(r'C:\Users\Tasnia Wahid\Downloads\cleaned_heart_data (2).csv')

# Define the colorise function
def colorise(value, threshold=0.6, equality='greater', absolute=True, font='red'):
    """
    Apply a style to a value based on a threshold.

    :param value: The value to be styled.
    :param threshold: The threshold for applying the style.
    :param equality: Determines if the style is applied for values greater than or less than the threshold.
    :param absolute: Consider the absolute value or not.
    :param font: The font color to apply.
    :return: A string representing the CSS to apply to the value.
    """
    apply_style = False
    if absolute:
        value_to_check = abs(value)
    else:
        value_to_check = value
    
    if equality == 'greater' and value_to_check > threshold:
        apply_style = True
    elif equality == 'less' and value_to_check < threshold:
        apply_style = True
    
    if apply_style:
        return f'color: {font};'
    else:
        return ''

# Now apply the styling to the correlation matrix
styled_corr = df.corr().style.applymap(colorise, threshold=0.6, equality='greater', absolute=True, font='red')
styled_corr
```
